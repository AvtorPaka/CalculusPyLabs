### Задание 1
Изобразить на рисунках область определения функций

1) $z=\ln(x^2+y)$

2) $z=\sqrt{\sin(x^2+y^2)}$

```python
x, y = sympy.symbols('x y', real = True)
graph_1 = (x**2 + y > 0)
graph_2 = (Sin(x**2 + y**2) > 0)
lst_graphs = [(x**2 + y > 0),(Sin(x**2 + y**2) > 0), sympy.And((x**2 - 4 > 0), (9 - y**2 > 0))]
for i in range(len(lst_graphs)):
  display(Latex(fr"[{latex(i + 1)}]"))
  plot_implicit(lst_graphs[i], (x, -8, 8), (y, -8, 8), aspect_ratio = (1, 1), title = f"${sympy.latex(lst_graphs[i])}$")
```

### Задача 2.
Построить на отдельных графиках линии уровня функций $z=\ln(x^2+y)$ и $z=x^2-y^2$.

```python
display(Latex(fr"\text{{Непонятно как находить линии уровня, в листке семинара просто подставляются заведомо правильные значения C :D}}"))

C = sympy.Symbol('C', real = True)
def f_1(x, y):
  return sympy.log(x**2 + y)
def f_2(x, y):
  return x**2 - y**2

def contourline(func, x, y, C):
 return sympy.Eq(func(x, y), C)

lst_contourline_data_1 = [(-2, 'blue'), (-1, 'green'), (0, 'red'), (2, 'black'), (4, 'pink')]
lst_contourline_data_2 = [(-4, 'blue'), (-1, 'green'), (0, 'red'), (2, 'black'), (4, 'pink')]

display(Latex(fr"[1]."))
plt_1 = plot(ylabel = r"$y$",show = False)
for i in range(len(lst_contourline_data_1)):
  c, color_cur = lst_contourline_data_1[i]
  plt_1.append(plot_implicit(contourline(f_1, x, y, c), line_color = color_cur, line_width = 20, show = False)[0])

plt_1.show()

display(Latex(fr"[2]."))
plt_2 = plot(ylabel = r"$y$", show = False)
for i in range(len(lst_contourline_data_2)):
  c, color_cur = lst_contourline_data_2[i]
  plt_2.append(plot_implicit(contourline(f_2, x, y, c), line_color = color_cur, line_width = 20, show = False)[0])

plt_2.show()
```

### Задача 3.
Найти и изобразить поверхности уровня функции $u=(x-1)^2+(y-2)^2+z^2$.

```python
display(Latex(fr"\text{{Непонятно как находить поверхности уровня, в листке семинара просто подставляются заведомо правильные значения C :D}}"))
z, c = sympy.symbols('z c', reals = True)
def f_3(x, y, z):
  return (x - 1)**2 + (y - 2)**2 + z**2
z_some = solve(sympy.Eq(f_3(x, y, z), c), z)
display(*z_some)
plot3d(z_some[0].subs({c: 4}), z_some[1].subs({c : 4}))
plot3d(z_some[0].subs({c: 2}), z_some[1].subs({c: 2}))
```

### Задача 6.
Проверить выполнение теоремы Эйлера для функции $f(x,y)=\frac{x}{x^2+y^2} $.
###### Указание.
Упростить выражение с помощью simplify, прежде чем сравнивать.

```python
k = sympy.Symbol('k', real = True)
def f_6(x, y):
  return sympy.S(x)/(x**2 + y**2).simplify()

display(Latex(fr"\text{{Функция является однордной так как }} \frac{{f(xk, yk)}}{{f(x, y)}} \ = {latex(f_6(x * k, y * k)/f_6(x, y))} \ = k^{{-1}}"))

display(Latex(fr"\text{{Теорема Эйлера для функции }} {latex(f_6(x, y))} {'не' * ((x * sympy.simplify(f_6(x, y).diff(x) * x + y * f_6(x,y).diff(y))) == 0)} \text{{ выполняется}}"))
```
