### Задание 1
Вычислить объем тела вращения около оси ОX фигуры, ограниченной графиком $y = \sin x$, $0 \le x \le \pi$ и осью ОХ. Построить график с помощью plot из sympy, названием графика (title) сделать формулу для вычисления объема по образцу Примера 1, например
$$
\pi\int\limits_0^1 (x - x^2)^2\,dx = \frac{\pi}{30},
$$
только формула должна соответствовать Заданию 1. Использовать latex, mode='inline'.

```python
x,y = symbols('x y')
def f(x):
  return sympy.sin(x)
Int_1 = sympy.pi * Integral(f(x)**2, (x, 0 , sympy.pi))
display(Latex(fr"\text{{V тела вращения = }} {latex(Int_1)}\ =\  {latex(Int_1.doit())}"))
plot_implicit(sympy.And(y < f(x), x < sympy.pi, y > 0), (x, -0.5, sympy.pi), (y, -0.5, 1.5), xlim = (0, sympy.pi), ylim = (-0.5, 1.5), title = sympy.latex(Int_1, mode = "inline"))
```

### Задание 2.
Вычислить объем тела, получающегося вращением вокруг ОХ астроиды
$$
\left\{
\begin{matrix}
x = \cos^3 t\\
y = \sin^3 t
\end{matrix}
\right.
$$

###### Указание.
Пределы интегрирования $\pi, 0$, поскольку $\cos^3 t$ убывает на $[0, \pi]$.

```python
t = Symbol('t')
def x_parametric(t):
  return sympy.cos(t)**3
def y_parametric(t):
  return sympy.sin(t)**3
Int_2 = sympy.pi * Integral((y_parametric(t)**2 * x_parametric(t).diff(t)), (t, sympy.pi, 0))
display(Latex(fr"\text{{V тела вращения = }} {latex(Int_2)}\ =\  {latex(Int_2.doit())}"))
```

### Задание 3.

Построить график при $a = 1$ в полярной системе координат, воспользоваться функцией polar из matplotlib.pyplot.

```python
phi, a = symbols('phi, a')
Phi = np.linspace(0, np.pi, 200)
plt.polar(Phi, 1 * np.cos(Phi)**2, color = "green", lw = 2)
def curve(a, phi):
  return a * sympy.cos(phi)**2
Int_3 = S(2)/3 * (sympy.pi * Integral(curve(a, phi)**3 * sympy.sin(phi), (phi, 0, sympy.pi)))
display(Latex(fr"\text{{V тела вращения = }} {latex(Int_3)}\ =\  {latex(Int_3.doit())}"))
print()
```

### Индивидуальное задание
Вычислить объем тела, образованного при вращении области между графиками функций
$f_1(x)$ и  $f_2(x)$.


Вариант №134

$$f(x)=2 \cos{\left(x \right)}$$
$$g(x)=2 \cos{\left(5 x \right)}$$


```python
def f(x):
  return 2 * sympy.cos(x)
def g(x):
  return 2 * sympy.cos(5 * x)
lst_ans_first_p = sorted([*sympy.solveset(sympy.Eq(f(x), g(x)), domain = sympy.Interval(0, 2 * sympy.pi))])[1]
Int_ind = sympy.pi * Integral((f(x)**2 - g(x)**2), (x, 0, lst_ans_first_p))
display(Latex(fr"\text{{V тела вращения = }} {latex(Int_ind)}\ =\  {latex(Int_ind.doit())}"))
```
